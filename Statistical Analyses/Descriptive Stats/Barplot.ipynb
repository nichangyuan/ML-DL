```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ebola.csv",encoding ='latin-1')
```

```python
df.head(10)
```

```python
sns.barplot(x="Type", y="Death", hue="Countries", data=df, ci=None)
plt.title("Ebola mortality distribution \n across Sierra Leone and Guinea Conakry")
plt.xlabel("Type of causes")
plt.ylabel("Number of ebola deaths")
```

```python
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax = sns.barplot(x="Type", y="Death", hue="Countries", data=df, ci=None)
ax.set(xlabel = 'Type of causes', ylabel = 'Number of Ebloa Death', )
ax.legend(loc=5)

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .4)

change_width(ax, .4)
```
