```python
import pandas as pd
```

```python
df = pd.read_csv('BaseballPlayers.csv')
```

```python
df.head(20)
```

### Objective: Summarize baseball players data by position and: <br>
> 1. Count: 'count'
> 2. Min:   'min'
> 3. Max:   'max'
> 4. Mean:  'mean'
> 5. Median: 'median'
> 6. std:    'std'
> 7. mad:    'mad'
> 8. skew:   'skew'

```python
bygroup_team = df.groupby(['Position'])['Weight']
bygroup_team
```

```python
df2 = bygroup_team.aggregate(['count', 'min', 'max', 'mean', 'median', 'std', 'mad', 'skew']).round(1)
```

```python
df2.columns = ['Count', 'Min', 'Max', 'Mean', 'Median', 'std', 'MAD', 'Skewness']
```
