```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data source. [Uniform Crime reporting statistics:](https://www.ucrdatatool.gov/Search/Crime/State/StatebyState.cfm) <br>  Violent crime in Washington DC

```python
df = pd.read_csv("WashingtonDCCrime.csv",encoding ='latin-1')
```

```python
df.head()
```

```python
Summary = pd.crosstab(df['Year'], df['Crime'], values=df['Rate'], aggfunc=np.sum)
```

```python
Summary
```

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})
Summary.plot(kind="area",stacked=True)
plt.title("Stacked area plot of Washington DC violent crime data")
plt.ylabel("Crime Rate per 100000s")
plt.xlabel("Year reported")
plt.savefig('stacked_area.png', dpi=1200)
plt.show()
```
