```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CPI-2014.csv",encoding ='latin-1')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4))
sns.set_context("notebook", font_scale=1.5, rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})
sns.boxplot(x='Continent', y = 'CPI_2014', data=df)
plt.title("Boxplot distribution of CPI 2014 data")
plt.xlabel("World Continents")
plt.ylabel("Corruption Perception Scores")
plt.xticks(rotation=20)
plt.savefig('corruption.png', dpi=1200)
plt.show()
```

#### 

```python

fig, ax = plt.subplots(1,1,figsize=(8,4))

sns.set_context("notebook", font_scale=1.5, rc={"font.size":14,"axes.titlesize":12,"axes.labelsize":12})
ax = sns.boxplot(x='Continent', y = 'CPI_2014', data=df)
ax.set_title("Boxplot distribution of CPI 2014 data")
ax.set_xlabel("World Continents")
ax.set_ylabel("Corruption Perception Scores")
ax.set_xticklabels(ax.get_xticklabels(), rotation=20)
fig.savefig('corruption.png', dpi=1200)
plt.show()
```
