```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Duncan.csv",encoding ='latin-1')
```

```python
df.head()
```

## Sort data by descending order of prestige score

```python
df = df.sort_values('Prestige',ascending=False).reset_index(drop=True)
```

```python
df.head()
```

```python
plt.rcParams["figure.figsize"] = [12,15]
```

```python
sns.set_style("darkgrid")
```

```python
sns.barplot(x="Prestige", y="Title", data=df, color="b")
plt.title("Duncan's occupation data analysis")
plt.ylabel("Professions titles")
plt.xlabel("Prestige scores")
plt.show()
```
