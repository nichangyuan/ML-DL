```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Bloodpressure.csv",encoding ='latin-1')
```

```python
df.head(10)
```

## Define figure size and default tilte and axes labels font sizes

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16})
sns.set_style("darkgrid")
```

```python
sns.regplot(x="Age", y="SBP", fit_reg=True, data=df)
plt.title("Relationship between Age and SBP")
plt.xlabel("Age of the patients")
plt.ylabel("SBP of patients")
plt.show()
```

```python
df2 = df[df['SBP'] != 220 ]
```

```python
sns.regplot(x="Age", y="SBP", fit_reg=True, data=df2)
plt.title("Relationship between Age and SBP")
plt.xlabel("Age of the patients")
plt.ylabel("SBP of patients")
plt.show()
```
