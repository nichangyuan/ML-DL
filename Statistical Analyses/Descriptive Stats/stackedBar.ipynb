```python
df = pd.read_csv("Migrants.csv",encoding ='latin-1')
```

```python
df.head()
```

### Subset the migrants deaths to certain regions

```python
df2 = df.loc[df['Region_of_Incident'].isin(['Horn of Africa','Mediterranean','Middle East','North Africa','Sub-Saharan Africa'])]
```

### Create a crosstab tables with the total deaths and missings

```python
Summary = pd.crosstab(df2['Region_of_Incident'], df2['Reported_Year'], values=df2['Total_Dead_and_Missing'],aggfunc=np.sum)
```

```python
Summary
```

### Use seaborn to style and pandas to plot the bar graphs

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
plt.rcParams["figure.figsize"] = [10,6]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16})
sns.set_style("darkgrid")
Summary.plot(kind="bar", stacked=False,color=flatui)
plt.title('Number of deaths or missing African migrants \n Period of 2014 thru 2017')
plt.ylabel('Number of deaths or missing migrants')
plt.xlabel('Region of casualties')
plt.xticks(rotation=-45)
plt.show()
```
