```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("LifeExpectancy.csv",encoding ='latin-1')
```

```python
df.head()
```

```python
plt.rcParams["figure.figsize"] = [12,8]
sns.set_style("darkgrid")
colors = ['#80b1d3','#bebada']
df.plot(x="Countries", y = ["Males", "Females"])
plt.title("Comparing males and females life expectancy in the world")
plt.xlabel("Countries of the world")
plt.ylabel("Life expectancy")
plt.xticks(df.index,df['Countries'],rotation=90)
plt.show()
```
