## Testing using Conover after Friedman chi-square

### Post-hoc pairwise test for multiple comparisons of mean rank sums (Nemenyi’s test) 
####  posthoc_conover(a, val_col=None, group_col=None,p_adjust = 'str', sort=True)
1. ‘bonferroni’ : 
2. One-step correction ‘sidak’
3. one-step correction ‘holm-sidak’ 
4. step-down method using Sidak adjustments ‘holm’ 
5. step-down method using Bonferroni adjustments ‘simes-hochberg’ 
6. step-up method (independent) ‘hommel’ 
7. closed method based on Simes tests (non-negative) ‘fdr_bh’ 
8. Benjamini/Hochberg (non-negative) ‘fdr_by’ 
9. Benjamini/Yekutieli (negative) ‘fdr_tsbh’ 
10. two stage fdr correction (non-negative) ‘fdr_tsbky’ 

### Use Pandas float format to display the p-values in float with 5 decimals places

```python
pd.options.display.float_format = '{:,.5f}'.format
```

```python
sp.posthoc_conover(df_melt, val_col='Scores', group_col='Treatment',p_adjust = 'bonferroni')
```

```python
sp.posthoc_nemenyi(df_melt, val_col='Scores', group_col='Treatment',dist="chi")
```
