```python
from scipy.stats import normaltest
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Test if the weights of goats are normally distributed<br> Ho: Distribution is Normal<br>H1: Distribution is not Normal

```python
df1 = pd.read_csv("C:/Users/ouaga/Dropbox/PythonScripts/Stats_tests/Goats50.csv")
```

```python
df1.head(10)
```

```python
normaltest(df1['Method1'])
```

```python
plt.rcParams["figure.figsize"] = [10,6]
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16}) 

sns.distplot(df1['Method1'],fit=stats.norm,kde=False)
plt.title('Histogram distribution of goats weights')
plt.xlabel('Weights of goats using a special type of feeds')
plt.show()
```

## Test if the weights of baseball players are normally distributed<br> Ho: Distribution is Normal<br>H1: Distribution is not Normal

```python
df2 = pd.read_csv("C:/Users/ouaga/Dropbox/PythonScripts/Stats_tests/BaseballPlayers.csv")
df2 = df2.dropna()
```

```python
df2.head()
```

```python
stats.normaltest(df2['Weight'])
```

```python
plt.rcParams["figure.figsize"] = [10,6]
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16}) 

sns.distplot(df2['Weight'],fit=stats.norm,kde=False)
plt.title('Histogram distribution of baseball players weights')
plt.xlabel('Weights of baseball players')
plt.show()
```

## Test if the Age of baseball players are normally distributed

```python
stats.normaltest(df2['Age'])
```

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16}) 

sns.distplot(df2['Age'],fit=stats.norm,kde=False)
plt.title('Histogram distribution of baseball players age')
plt.xlabel('Age of baseball players')
plt.show()
```
